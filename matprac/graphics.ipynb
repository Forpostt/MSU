```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(15, 8))
with open("res.txt") as fin:
    for i, line in enumerate(fin):
        y = map(float, line.split())
        plt.plot(np.arange(0,100, 5), y[:0:-1], label="size= {}".format(int(y[0])))

plt.xlabel("percentage of edges from size")
plt.ylabel("time of solution")
plt.legend()
```
